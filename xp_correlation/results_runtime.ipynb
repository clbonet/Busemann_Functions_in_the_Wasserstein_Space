```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_dataset_size = [5000 * (i + 1) for i in range(int(60000 // 5000))]
```

```python
str_dataset = "MNIST"
path_results = "./runtime_corr_" + str_dataset
n_try = 100
n_projs = 1000

mean_times_otdd, std_times_otdd = [], []
mean_times_sotdd, std_times_sotdd = [], []
mean_times_swb1dg, std_times_swb1dg = [], []
mean_times_swbg, std_times_swbg = [], []
mean_times_my_sotdd, std_times_my_sotdd = [], []

for dataset_size in list_dataset_size:
    save_dir = path_results + f"/{str_dataset}_try_{n_try}_size_{dataset_size}"

    # try:
    #     L_otdd = np.loadtxt(f'{save_dir}/otdd_runtime.txt', delimiter=",")
    #     mean_times_otdd.append(np.mean(L_otdd))
    #     std_times_otdd.append(np.std(L_otdd))
    # except:
    #     mean_times_otdd.append(np.inf)
    #     std_times_otdd.append(np.inf)

    try:
        L_sotdd = np.loadtxt(f'{save_dir}/sotdd_{n_projs}_runtime.txt', delimiter=",")
        mean_times_sotdd.append(np.mean(L_sotdd[1:]))
        std_times_sotdd.append(np.std(L_sotdd[1:]))
    except:
        mean_times_sotdd.append(np.inf)
        std_times_sotdd.append(np.inf)

    try:
        L_swb1dg = np.loadtxt(f'{save_dir}/swb1dg_{n_projs}_runtime.txt', delimiter=",")
        mean_times_swb1dg.append(np.mean(L_swb1dg[1:]))
        std_times_swb1dg.append(np.std(L_swb1dg[1:]))
    except:
        mean_times_swb1dg.append(np.inf)
        std_times_swb1dg.append(np.inf)

    try:
        L_swbg = np.loadtxt(f'{save_dir}/swbg_{n_projs}_runtime.txt', delimiter=",")
        mean_times_swbg.append(np.mean(L_swbg[1:]))
        std_times_swbg.append(np.std(L_swbg[1:]))
    except:
        mean_times_swbg.append(np.inf)
        std_times_swbg.append(np.inf)

    try:
        L_my_sotdd = np.loadtxt(f'{save_dir}/my_sotdd_{n_projs}_runtime.txt', delimiter=",")
        mean_times_my_sotdd.append(np.mean(L_my_sotdd[1:]))
        std_times_my_sotdd.append(np.std(L_my_sotdd[1:]))
    except:
        mean_times_my_sotdd.append(np.inf)
        std_times_my_sotdd.append(np.inf)

    # print(dataset_size, mean_times_otdd[-1], mean_times_sotdd[-1], mean_times_swb1dg[-1], mean_times_swbg[-1])

mean_times_otdd, std_times_otdd = np.array(mean_times_otdd), np.array(std_times_otdd)
mean_times_sotdd, std_times_sotdd = np.array(mean_times_sotdd), np.array(std_times_sotdd)
mean_times_swb1dg, std_times_swb1dg = np.array(mean_times_swb1dg), np.array(std_times_swb1dg)
mean_times_swbg, std_times_swbg = np.array(mean_times_swbg), np.array(std_times_swbg)
mean_times_my_sotdd, std_times_my_sotdd = np.array(mean_times_my_sotdd), np.array(std_times_my_sotdd)
```

```python
# plt.plot(list_dataset_size, mean_times_otdd, label="OTDD")
# plt.fill_between(list_dataset_size, mean_times_otdd-std_times_otdd, mean_times_otdd+std_times_otdd,alpha=0.5)

# plt.plot(list_dataset_size, mean_times_sotdd, label="sOTDD")
# plt.fill_between(list_dataset_size, mean_times_sotdd-std_times_sotdd, mean_times_sotdd+std_times_sotdd,alpha=0.5)

plt.plot(list_dataset_size, mean_times_swb1dg, label="SWB1DG")
plt.fill_between(list_dataset_size, mean_times_swb1dg-std_times_swb1dg, mean_times_swb1dg+std_times_swb1dg,alpha=0.5)

plt.plot(list_dataset_size, mean_times_swbg, label="SWBG")
plt.fill_between(list_dataset_size, mean_times_swbg-std_times_swbg, mean_times_swbg+std_times_swbg,alpha=0.5)

plt.plot(list_dataset_size, mean_times_my_sotdd, label="SOTDD")
plt.fill_between(list_dataset_size, mean_times_my_sotdd-std_times_my_sotdd, mean_times_my_sotdd+std_times_my_sotdd,alpha=0.5)

plt.grid(True)

plt.legend()
plt.show()
```
