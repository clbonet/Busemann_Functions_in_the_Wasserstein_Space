```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L_w2_sotdd = np.loadtxt("./results/convergence_sotdd.csv", delimiter=",")
L_w2_swb1dg = np.loadtxt("./results/convergence_swb1dg.csv", delimiter=",")
L_w2_swbg = np.loadtxt("./results/convergence_swbg.csv", delimiter=",")
```

```python
mean_sotdd = np.mean(L_w2_sotdd, axis=0)
std_sotdd = np.std(L_w2_sotdd, axis=0)

mean_swb1dg = np.mean(L_w2_swb1dg, axis=0)
std_swb1dg = np.std(L_w2_swb1dg, axis=0)

mean_swbg = np.mean(L_w2_swbg, axis=0)
std_swbg = np.std(L_w2_swbg, axis=0)
```

```python
plt.figure(figsize=(6, 3))
# plt.figure(figsize=(5,2))

plt.plot(np.arange(1001), mean_sotdd, label="SOTDD")
plt.fill_between(np.arange(1001), mean_sotdd-std_sotdd, mean_sotdd+std_sotdd,alpha=0.5)

plt.plot(np.arange(1001), mean_swb1dg, label="SWB1DG")
plt.fill_between(np.arange(1001), mean_swb1dg-std_swb1dg, mean_swb1dg+std_swb1dg,alpha=0.5)

plt.plot(np.arange(1001), mean_swbg, label="SWBG")
plt.fill_between(np.arange(1001), mean_swbg-std_swbg, mean_swbg+std_swbg,alpha=0.5)

plt.ylabel(r"$\mathrm{W}_{\mathrm{W}_2}$", usetex=True, fontsize=13, rotation=0)
plt.xlabel(r"Time ($\tau \cdot$ iter)", fontsize=13, usetex=True)
plt.yscale("log")
plt.grid(True)
plt.legend(fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(6, 2))  # compact size

# Plot SOTDD
plt.plot(np.arange(1001), mean_sotdd, label="SOTDD")
plt.fill_between(
    np.arange(1001),
    mean_sotdd - std_sotdd,
    mean_sotdd + std_sotdd,
    alpha=0.2
)

# Plot SWB1DG
plt.plot(np.arange(1001), mean_swb1dg, label="SWB1DG")
plt.fill_between(
    np.arange(1001),
    mean_swb1dg - std_swb1dg,
    mean_swb1dg + std_swb1dg,
    alpha=0.2
)

# Plot SWBG
plt.plot(np.arange(1001), mean_swbg, label="SWBG")
plt.fill_between(
    np.arange(1001),
    mean_swbg - std_swbg,
    mean_swbg + std_swbg,
    alpha=0.2
)

# Labels and styling
plt.ylabel(
    r"$\mathrm{W}_{\mathrm{W}_2}^2$",
    usetex=True,
    fontsize=14,
    rotation=0,
    labelpad=20  # much larger padding so its readable
)
plt.xlabel(r"Time ($\tau \cdot$ iter)", fontsize=14, usetex=True)

plt.yscale("log")
plt.grid(True, which="both", ls="--", lw=0.5, alpha=0.7)
plt.yticks([1,1e-2,1e-4,1e-6,1e-8])
plt.legend(fontsize=11, loc="best", frameon=True, ncols=3)

plt.tight_layout()

plt.savefig("./Convergence_gfs_rings.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
